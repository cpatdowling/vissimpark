```python
import xml.etree.ElementTree as ET
import numpy as np
import os
import sys
import shutil
```

```python
# Here we'll generate multiple directories with basic configuration files copied, and the main configuration 
# and the main configuration file containing static volumes, which a unique volume for each directory

write_directory = "C:\\Users\chase\\Documents\\vissimpark\\VISSIM_configs\\baseline prototype\\single_volume_runs"
base_config_directory = "C:\\Users\\chase\\Documents\\vissimpark\\VISSIM_configs\\baseline prototype\\base_config"

os.listdir(base_config_directory)

#increments of 5000 down from 100000 to 5000
#copy the base directory, edit the prototype.ipnx file
```

```python
volumes = np.arange(5000, 105000, 5000)
```

```python
#write experiment directories, copy files, and edit volumes

basefiles = os.listdir(base_config_directory)

for vol in volumes:
    outdir = write_directory + "\\net_vol_" + str(vol)
    if not os.path.exists(outdir):
        os.mkdir(outdir)
    
    for file_name in basefiles:
        full_file_name = os.path.join(base_config_directory, file_name)
        if os.path.isfile(full_file_name):
            shutil.copy(full_file_name, outdir)
            
    tree = ET.parse(base_config_directory + "\\prototype.inpx")
    root = tree.getroot()
    for v in root.iter('timeIntervalVehVolume'):
        v.set('volume', str(vol))
    
    tree.write(outdir + "\\prototype.inpx")
```
